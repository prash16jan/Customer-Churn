```python
# Converting TotalCharges to Numeric data type
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors='coerce')
```

```python
# Removing NA Values
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# Removing Customer ID in the Data Frame
```

```python
df2 = df.iloc[:,1:]
df2.head()
```

```python
# Converting Target Variable to Numeric Variable
```

```python
df2['Churn'].replace(to_replace='Yes', value=1, inplace=True)
df2['Churn'].replace(to_replace='No', value=0, inplace=True)
```

```python
df2.head()
```

```python
# Converting All Categoricall variable into Dummies
```

```python
df_dum = pd.get_dummies(df2)
df_dum.head()
```

```python
df_dum.corr()
```

```python
# Splitting The Data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_dum.drop("Churn", axis=1)
y = df_dum["Churn"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```
